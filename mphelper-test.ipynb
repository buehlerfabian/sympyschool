```python
import matplotlib.pyplot as plt
import numpy as np
import sympyschool.mplhelper as mh
```

```python
mh.set_usetex(True)
```

```python
x = np.linspace(0,2*np.pi,100)
fig, ax = plt.subplots()
mh.sbplot(x, np.sin(x), color='r', linewidth=2, xlabel='was langes')
ax.grid(which='both')
mh.set_pi_axis_formatter(ax.xaxis)
mh.set_xticks(np.pi/4, ax)
mh.set_yticks(.5)
ax.set_aspect("equal")
mh.set_minor_xticks(2)
```
